## Library imports

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

## Settings

```python
pd.plotting.register_matplotlib_converters()
pd.set_option('display.max_colwidth',  None)
%matplotlib inline
```

## Dataset import

```python
filepath = '../datasets/iris dataset/iris.csv'
iris = pd.read_csv(filepath)
iris.head()
```

```python
filepath = '../datasets/wine reviews/winemag-data-130k-v2.csv'
wine_reviews = pd.read_csv(filepath)
wine_reviews.head()
```

# Matplotlib 

## Scatter plot

```python
#Creating figure and axis
fig, ax = plt.subplots(figsize = (14, 6))

#Scatter speal_length against sepal_width
ax.scatter(iris['sepal_length'], iris['sepal_width'])

#Set title and axis labels
ax.set_title('Iris dataset')
ax.set_xlabel('Sepal length')
ax.set_ylabel('Sepal width')
plt.show()
```

```python
#Classes
for species in iris['species'].unique():
    print(species)
```

```python
#Coloring according to species
colors = {'Iris-setosa': 'r', 'Iris-versicolor': 'g', 'Iris-virginica': 'b'}

fig, ax = plt.subplots(figsize = (14, 6))

for i in range(len(iris['sepal_length'])):
    ax.scatter(iris['sepal_length'][i], iris['sepal_width'][i], color = colors[iris['species'][i]])
    
ax.set_title('Iris dataset with colours')
ax.set_xlabel('Sepal length')
ax.set_ylabel('Sepal width')
plt.show()
```

## Line chart

```python
#Get columns to plot
columns = iris.columns.drop(['species'])

x_data = range(0, iris.shape[0])

fig, ax = plt.subplots(figsize = (17, 8))

for column in columns:
    ax.plot(x_data, iris[column], label = column)
    
ax.set_title('Iris dataset line chart', fontsize = 14)
ax.set_xlabel('flower no', fontsize = 14)
ax.set_ylabel('length/width', fontsize = 14)
ax.legend()
plt.show()
```

## Histogram

```python
fig, ax = plt.subplots(figsize = (16, 8))

ax.hist(wine_reviews['points'])

ax.set_title('Wine review dataset histogram', fontsize = 14)
ax.set_xlabel('Points', fontsize = 14)
ax.set_ylabel('Frequency', fontsize = 14)
plt.show()
```

## Bar graph

```python
#Create figure and axis
fig, ax = plt.subplots(figsize = (16, 8))

#Count the occurance of each class
data = wine_reviews['points'].value_counts()

#Get x and y avlues
points = data.index
frequency = data.values

#Create bar chart
ax.bar(points, frequency)

ax.set_title('Wine review scores', fontsize = 14)
ax.set_xlabel('Points', fontsize = 14)
ax.set_ylabel('Frequency', fontsize = 14)
plt.show()
```

## Bar graph of average wine prices by country

```python
wine_reviews.groupby('country').price.mean().sort_values(ascending = False)[:10].plot.bar(ylabel = 'Average price', 
                                                                                          title = 'Average wine prices')
plt.show()
```

## Heatmap

```python
corr = iris.corr()
fig, ax = plt.subplots()

#Creating heatmap
im = ax.imshow(corr.values)

ax.set_xticks(np.arange(len(corr.columns)))
ax.set_yticks(np.arange(len(corr.columns)))
ax.set_xticklabels(corr.columns)
ax.set_yticklabels(corr.columns)

plt.setp(ax.get_xticklabels(), rotation = 45, ha = 'right', rotation_mode = 'anchor')
plt.show()
```

```python
corr = iris.corr()
fig, ax = plt.subplots()

#Creating heatmap
im = ax.imshow(corr.values)

ax.set_xticks(np.arange(len(corr.columns)))
ax.set_yticks(np.arange(len(corr.columns)))
ax.set_xticklabels(corr.columns)
ax.set_yticklabels(corr.columns)

plt.setp(ax.get_xticklabels(), rotation = 45, ha = 'right', rotation_mode = 'anchor')

# Loop over data dimensions and create text annotations.
for i in range(len(corr.columns)):
    for j in range(len(corr.columns)):
        text = ax.text(j, i, np.around(corr.iloc[i, j], decimals=2),
                       ha="center", va="center", color="black")
```
